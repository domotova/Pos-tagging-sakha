```
with open('/content/корпус-2.txt') as f:
    s = f.read()
    for ss in '.,!»?—–-:)”;':
        s = s.replace(ss, ' '+ss)
    for ss in '(“«—–-':
        s = s.replace(ss, ss+' ')
    corpus = s.split('\n')
```

```
with open('/content/pos-tag.txt') as f:
    tags = f.read().split('\n')
```

```
print(len(corpus), len(tags))
```

```
sentences = []
tag = []
for i in range(len(tags)):
    sentences.append(corpus[i].split(' ')[1:])
    tag.append(tags[i].split(' '))
```

```
en_tag = {
	1 : 'NOUN', 
	2 : 'ADJ', 
	3 : 'NUM', 
	4 : 'PRON', 
	5 : 'VERB', 
	6 : 'ADV', 
	8 : 'PR', 
	10 : 'CONJ', 
	7 : 'PART', 
  9 : 'AUX',
	11 :'INTJ', 
	12 : 'PUNCT', 
  13 : 'SYM',
  14 : 'X'
}
```

```
tagged = []
for i in range(len(tags)):
  ll = []
  for j in range(len(sentences[i])):
	    ll.append([sentences[i][j], en_tag[int(tag[i][j])]])
  tagged.append(ll)
```

```
from conllu import TokenList
from collections import OrderedDict
```

```
def compile_sent(sent):
    sent_list = TokenList()
    for i, tok_data in enumerate(sent):
        tok_id = i + 1
        tok = tok_data[0]
        pos = tok_data[1]
        compiled_tok = OrderedDict({'id': tok_id, 'form': tok, 'lemma': None, 'upostag': pos, 'xpostag': None, 'feats': None, 'head': None, 'deprel': None, 'deps': None, 'misc': None})
        sent_list.append(compiled_tok)
    sent_list = sent_list.serialize()
    return sent_list
```

```
with open('supervised_tag.connlu', 'w') as f:   
  for i in range(len(tagged)):
    f.write('# sent_id = '+str(i)+'\n')
    f.write('# text = '+corpus[i]+'\n')
    f.write(compile_sent(tagged[i]))
    f.write('\n')
```
